Iterators are created using the iter() function, and they can be iterated over using the next() function.

```python
def test3(n):
  a,b = 0,1
  for i in range(0,n):
    yield a
    a,b = b, a+b
```

```python
test3(10)
```

```python
for i in test3(10):
  print((i))
```

What is a generator function in python? Why yield keyword is used? Give an example of a generator
function.

Generator function is used when u dont want the function to do the process as a whole and instead generate all the process in the function one by one.
Yield keyword is used to return the value. The above fibonacci problem is an example for generator function.

```python
#Q5. Create a generator function for prime numbers less than 1000. Use the next() method to print the
# first 20 prime numbers.
def test_prime():
  l = []
  for i in range(2,1000):
    for j in range(2,i):
      if i%j==0:
        break
    else:
      l.append(i)
      yield i
```

```python
generator = test_prime()
for i in range(20):
    prime = next(generator)
    print(prime)
```

```python
# Q6. Write a python program to print the first 10 Fibonacci numbers using a while loop.

a,b = 0,1
n = 10
while n>0:
  print(a)
  a,b=b,a+b
  n-=1
```
